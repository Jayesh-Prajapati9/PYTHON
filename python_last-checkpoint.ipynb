```python
L=[[1,1,2,3,2],[4,3,2,1,3],[5,2,1,3,2],[3,2,1,4,5],[2,3,1,4,3]]

d={}
for i in L:
    for j in i:
        d[j]=d.get(j,0)+1
print(d)
L1=list(d.items())
print(L1)
L1.sort(key=lambda x:x[1])
print(L1)
```
