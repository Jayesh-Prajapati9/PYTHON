```python
# write a program that accepts a tuple from user and find minimum odd number,max odd number, min even number
# max even number from given tuple using only tuple
t1=eval(input("Enter The Tuple:- "))
minodd=None
maxodd=None
mineven=None
maxeven=None
for i in t1:
    if(i%2==0):
        if (maxeven==None or maxeven<i):
            maxeven=i
        if (mineven==None or mineven>i):
            mineven=i
    else:
        if (maxodd==None or maxodd<i):
            maxodd=i
        if (minodd==None or minodd>i):
            minodd=i
print("MinEven:-",mineven)
print("MaxEven:-",maxeven)
print("MinOdd:-",minodd)
print("MaxOdd:-",maxodd)
```
