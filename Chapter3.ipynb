```python
def greet():
    print("Hello World")

greet()
##by just writing the name without parenthesis we will get address of the function and with parenthesis it can be called, also this function returns none

def greet(name):
    print(f"Hello {name}")

greet("John")
###the string we pass while calling the function is the argument while the one written in the function is the parameter

def calc(radius:float) -> float:
    ##this input and output are just conventions, it's just type hinting
    """
    Description:It returns area of a circle
    Parameters: Radius as float
    Returns: Area of circle in float
    """
    return 3.14*radius*radius

calc()
##calc() missing 1 required positional argument: 'radius'

calc(5.5)

print(calc.__doc__)
##gives documentation written by us

help(calc)
##returns the documentation about the function

help(range)
##returns documentation on builtin functions

##Function Categories
def printline():
    print("Hello")   ##zero parameter,zero return

printline()

def printline(s):
    print(s)   ##one parameter,zero return

printline('Hello')

def printline():
    return 'Hello'  ##zero parameter,one return

printline()

def printline(s):
    return s     ##one parameter,one return

printline("Hello")

##Return Multiple Variables

def calc(a,b) :
    return a+b,a-b,a*b

##the default return type for a function returning multiple values is a tuple

calc(3,2)

##create a pair of functions to convert faherenheit to celsius and celsius to faherenheit with the user choice as an input
temp = float(input("Enter temp:"))
choice = input("Is it in celsius(Enter C) or in faherenheit(Enter F)?:")
def ftoc(temp):
    celsius = (5/9)*(temp - 32)
    print(f"{celsius:.2f} Celsius")
    
def ctof(temp) :
    faherenheit = ((9/5)*temp)+32
    print(f"{faherenheit:.2f} Faherenheit")
    
if choice == 'C':
    ctof(temp)
elif choice == 'F':
    ftoc(temp)
else:
    print('Invalid choice')

##Variable Scope and Lifetime
##written in cw

#1. Local scope
x = 90 #global scope
def example():
    x = 5   #local scope
    print(x)

example() #x = 5
print(x)  #x = 90

#Types of Arguments
#1. DEFAULT:
def square(a = 10) :
    return a*a

square() ##we have to write print here, if we don't print it the output will remain in the console
##O/P = 100

square(20) ##overrides the upper function
```

```python
#3.KEYWORD:
def wish(name,msg) :
    print('hello',name,msg)
```

```python
wish()
#wish() missing 2 required positional arguments: 'name' and 'msg'
```

```python
wish(name = 'hi')
#wish() missing 1 required positional argument: 'msg'
```

```python
wish(name = 'python',msg =  'hi')
```

```python
wish(msg = 'hi', name = 'java')
```

```python
wish('C++','Good Morning')
```

```python
wish(name= 'C',"Good Evening")
#positional argument follows keyword argument, which is not allowed
```

```python
wish('C',msg = 'Good Night')
#keyword argument can follow positional argument
```

```python
#4. VARIABLE LENGTH:
def add(a,b) :
    return a + b
def add(a,b) :
    return a - b
```

```python
add(3,2)
##abstraction is used here,the below function overrides the above one
```

```python
add(3,2,5,6,7)
#TypeError: add() takes 2 positional arguments but 5 were given
```

```python
def add(*n) :
    total = 0
    for i in n:
        total = total + i
    print(total)
    
# *n can take sequence or list
```

```python
add(1,2)
```

```python
add(1,2,3)
```

```python
##'global' KEYWORD:
x = 90
def func(a):
    global x #changes the global value
    x = 5
    return x * a
print(func(6)) #30
print(x) # if global not written then 90 else 5
```

```python
##RULES:
#1. Order of arguments in function defintion:
#1.1 positional arguments: they must be provided in order before any keyword arguments
#1.2 keyword arguments: they can be provided in any order after positional arguments
#1.3 Default arguments must appear after non default arguments
#1.4 *args collects additional positional arguments
```
