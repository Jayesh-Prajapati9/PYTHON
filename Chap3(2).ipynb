```python
##wap to check if a given number is perfect square or not
num = int(input("Enter choice:"))
square = False
for i in range(1,num) :
    if num == (i*i) :
        square = True
if square == True :
    print('Number is perfect square')
else :
    print('Number is not perfect square')
```
