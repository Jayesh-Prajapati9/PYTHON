```python
for i in range(5):
    for j in range(9) :
        if (i+j == 4) or (j-i == 4) or (i==2 and (j == 3 or j == 4 or j == 5)):
            print("*", end = " ")
        else :
            print(" ",end = " ")
    if i == 5:
        break
    print()
```

```python
for i in range(1,6) :
    for j in range(1,6) :
        if (i == j) or (j == ((5-i)+1)) :
            print("*", end = " ")
        else :
            print(" ",end = " ")
    print()         
```
