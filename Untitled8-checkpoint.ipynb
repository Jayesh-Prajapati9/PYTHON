```python
for i in range (1,5):
    for j in range (4,i-1,-1):
        print("*",end=" ")
    for k in range (1,i):
        print(" "*2,end="  ")
    for l in range (4,i-1,-1):
        print("*",end=" ")
    print("")
for i in range (1,5):
    for j in range (0,i):
        print("*",end=" ")
    for k in range (4,i,-1):
        print(" "*2,end="  ")
    for l in range (0,i):
        print("*",end=" ")
    print("")
```
