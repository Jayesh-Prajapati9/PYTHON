```python
##Collatz conjecture
##A collatz sequence is generated by applying following rules in turns
#1. if even divide by 2
#2. if odd multiply by 3 and add 1
#the algorithm says that, it always reaches 1 for all positive integers
##Create a function for two positive integers a and b and returns a if integer a took fewer steps to reach 1 than b and vice versa
a = int(input("Enter first number:"))
b = int(input("Enter second number:"))

def countSteps(num) :
    steps = 0
    while num!=1 :
        steps += 1
        if num%2 == 0:
            num = num//2
        else :
            num = (3*num) + 1
    return steps

def lesserSteps(a,b) :
    asteps = countSteps(a)
    bsteps = countSteps(b)
    if asteps < bsteps :
        return a
    else:
        return b
    
less = lesserSteps(a,b)
print(less)
```
